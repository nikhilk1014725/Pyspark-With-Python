### Fetching data from Databricks Delta Live Table (DLT)
This notebook demonstrates how to connect to a Databricks workspace and load data from a Delta Live Table.

```python
from pyspark.sql import SparkSession

# Configure SparkSession for Databricks.
# In Databricks this is already available. When running outside
# of Databricks, provide the necessary connection details.
spark = SparkSession.builder.appName("dlt-fetch-example").getOrCreate()

# Name of the Delta Live Table to query
table_name = 'LIVE.my_dlt_table'

df = spark.read.format('delta').table(table_name)
df.show()
```
